```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
df=pd.read_csv('sales_data_EDA_final.csv')
df
```

```python
x=df.drop(['Item_Outlet_Sales'],axis=1)
y=df['Item_Outlet_Sales']
```

```python
#Splitting the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape
```

```python
model=KNeighborsRegressor()
model.fit(x_train,y_train)
```

```python
#evaluate the model
#training_data
print('#'*50)
print('Training Data Evaluation')
print('#'*50)
y_pred=model.predict(x_train)

r2=r2_score(y_train,y_pred)
print('R2 Score==',r2)

mse=mean_squared_error(y_train,y_pred)
print('MSE==',mse)

mae=mean_absolute_error(y_train,y_pred)
print('MAE==',mae)

print('#'*50)
print('Testing Data Evaluation')
print('#'*50)

y_pred=model.predict(x_test)

r2=r2_score(y_test,y_pred)
print('R2 Score==',r2)

mse=mean_squared_error(y_test,y_pred)
print('MSE==',mse)

mae=mean_absolute_error(y_test,y_pred)
print('MAE==',mae)
```

```python
# Hyperparameter Tuning
model=KNeighborsRegressor()

hyp={'n_neighbors':np.arange(3,30),
     'p':[1,2]}

gscv_model=GridSearchCV(model,hyp,cv=5)
gscv_model.fit(x_train,y_train)
```

```python
gscv_model.best_estimator_
```

```python
model_01=gscv_model.best_estimator_
model_01.fit(x_train,y_train)
```

```python
#evaluate the model
#training_data
print('#'*50)
print('Training Data Evaluation')
print('#'*50)
y_pred=model_01.predict(x_train)

r2=r2_score(y_train,y_pred)
print('R2 Score==',r2)

mse=mean_squared_error(y_train,y_pred)
print('MSE==',mse)

mae=mean_absolute_error(y_train,y_pred)
print('MAE==',mae)

print('#'*50)
print('Testing Data Evaluation')
print('#'*50)

y_pred=model_01.predict(x_test)

r2=r2_score(y_test,y_pred)
print('R2 Score==',r2)

mse=mean_squared_error(y_test,y_pred)
print('MSE==',mse)

mae=mean_absolute_error(y_test,y_pred)
print('MAE==',mae)
```
